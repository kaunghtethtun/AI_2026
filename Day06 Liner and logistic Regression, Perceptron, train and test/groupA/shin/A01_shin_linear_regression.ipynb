```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load Dataset
df = pd.read_csv("heart.csv")

# Missing Value Remove
df = df[['age','sex']].dropna()

# Feature and sex
X = df[['age']]
y = df['sex']

# Model
model = LinearRegression()

model.fit(X,y)

# Prediction Line
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(8,6))

plt.scatter(X,y, alpha=0.5)

plt.plot(X,y_pred)

plt.xlabel("age")

plt.ylabel("sex")

plt.title("Linear Regression Prediction Line (House Price)")

plt.show()
```

```python
import pandas as pd

new_heart = pd.DataFrame({

    "age":[30]

})
predicted_age = model.predict(new_heart)

print("Predicted age :", predicted_age[0])
```

```python
new_heart = pd.DataFrame({

    "age":[30,40]

})

predictions = model.predict(new_heart)

print(predictions)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'],df['sex'])

plt.plot(X,model.predict(X))

# New house point
plt.scatter(new_heart,predictions,
            marker="x",
            s=200)

plt.show()
```
