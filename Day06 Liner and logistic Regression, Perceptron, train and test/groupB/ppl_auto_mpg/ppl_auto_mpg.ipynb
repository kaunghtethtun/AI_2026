```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'C:\Users\pyaep\OneDrive\Desktop\AI_2026\Day06 Liner and logistic Regression, Perceptron, train and test\groupB\ppl_auto_mpg\auto-mpg.csv')

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows where horsepower is NaN
df.dropna(subset=['horsepower'], inplace=True)

display(df.shape)

df.head()

print(df.dtypes)
```

```python
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]

y = df['mpg']
```

```python
# Linear Regression Model တည်ဆောက်မယ်
mlr_model = LinearRegression() # Model Object တစ်ခုကို စတင်ဖန်တီးခြင်း

# Model Training
mlr_model.fit(X, y)
```

```python
y_pred = mlr_model.predict(X)

print("First 10 Predictions:", y_pred[:10])
```

```python
print(mlr_model.coef_)
print(mlr_model.intercept_) 
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(y, y_pred, color='blue', alpha=0.6)

plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')

plt.xlabel('Actual MPG')  
plt.ylabel('Predicted MPG') 
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

accuracy = mlr_model.score(X, y) 

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

print(f"R2 Score: {accuracy:.4f}")
print(f"RMS Error: {rmse:.4f} MPG")
```
